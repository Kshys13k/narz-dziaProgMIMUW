```python
import pandas as pd
```

```python
leadername = ['Stanisław', 'Jan', 'Jan', 'Grzegorz', 'Jan']
leadersurname = ['Żółkiewski', 'Chodkiewicz', 'Sobieski', 'Chodkiewicz', 'Chodkiewicz']
battle = ['Kłuszyn 4.VII.1610', 'Kircholm 27.IX.1605', 'Wiedeń 12.IX.1683 | Chocim 11.XI.1673','Czaśniki 26.I.1564','Biały Kamień 25.IX.1604']
btype = ['land battle']*5
no = ['old slavic','hebrew','hebrew','greek','hebrew']
title = ['hetman wielki koronny','hetman wielki litewski','król','hetman wielki litewski','hetman wielki litewski']

ln = pd.Series(leadername)
ls = pd.Series(leadersurname)
b = pd.Series(battle)
bt = pd.Series(btype)
o = pd.Series(no)
t = pd.Series(title)

df = pd.DataFrame({'Leader Name': ln, 'Leader Surame': ls, 'Battle': b, 'Battle type': bt,
                     'Name origin': o, 'Title':t})
df
```

```python
#1
row=df.iloc[2]
battles=row["Battle"].split("|")
for battle in battles:
    row["Battle"]=battle
    df.loc[len(df)]=row

df=df.drop(2)
df=df.reset_index(drop=True)
df
```

```python
#2
df_leaders=df
df_leaders=df_leaders.drop_duplicates(subset=["Leader Name","Leader Surame"])
df_leaders=df_leaders.drop(columns=["Battle","Battle type"])
df_leaders=df_leaders.reset_index(drop=True)
df_leaders.index.name="Leader id"

df_battles=df
list_of_ids=[]
for index, row in df_battles.iterrows():
    leader=df_battles.loc[index,"Leader Surame"]
    id_=df_leaders.index[df_leaders["Leader Surame"]==leader].tolist()
    list_of_ids.append(id_[0])
new_df=pd.DataFrame({"Leader id":list_of_ids})
df_battles=pd.concat([df_battles, new_df],axis=1)
df_battles=df_battles.drop(columns=["Leader Name","Leader Surame", "Name origin", "Title"])
df_battles
```
